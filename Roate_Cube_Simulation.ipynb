```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def rotation_matrix(angle, axis):

    if axis == 'x':
        return np.array([[1, 0, 0],
                         [0, np.cos(angle), -np.sin(angle)],
                         [0, np.sin(angle), np.cos(angle)]])
    elif axis == 'y':
        return np.array([[np.cos(angle), 0, np.sin(angle)],
                         [0, 1, 0],
                         [-np.sin(angle), 0, np.cos(angle)]])
    elif axis == 'z':
        return np.array([[np.cos(angle), -np.sin(angle), 0],
                         [np.sin(angle), np.cos(angle), 0],
                         [0, 0, 1]])
    else:
        raise ValueError("Invalid axis. Please use 'x', 'y', or 'z'.")
```

```python
for i in range(3):
  # print("hii")

          # Define angle and axis of rotation
          angle = np.pi / (4)  # 45 degrees
          # axis = 'y'
          if i<1:
            axis = 'x'
          elif i<2:
            axis = 'y'
          else:
            axis = 'z'    
                  

          # Generate 3D coordinates
          x = np.linspace(-1, 1, 10) # -1 to 1 10 intervals on x axis
          y = np.linspace(-1, 1, 10)
          z = np.linspace(-1, 1, 10)

          X, Y, Z = np.meshgrid(x, y, z) 
          coords = np.vstack((X.flatten(), Y.flatten(), Z.flatten()))
          
          # Apply rotation matrix
          rot_matrix = rotation_matrix(angle, axis)
          rotated_coords = np.dot(rot_matrix, coords)

          # Plot original and rotated points
          fig = plt.figure()
          ax = fig.add_subplot(111, projection='3d')
          ax.scatter(coords[0], coords[1], coords[2], c='b', label='Original')
          ax.scatter(rotated_coords[0], rotated_coords[1], rotated_coords[2], c='r', label='Rotated')
          ax.set_xlabel('X')
          ax.set_ylabel('Y')
          ax.set_zlabel('Z')
          ax.legend()
          plt.show()



```
